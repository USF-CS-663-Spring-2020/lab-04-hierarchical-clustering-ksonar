```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
census = pd.read_csv('../USCensus1990-2000rows.data.txt')
census
```

```python
census_linkage = linkage(census, 'ward')
census_linkage.shape
```

```python
plt.figure(figsize=(20,15))
dendrogram(census_linkage, orientation='right')
plt.show()
```

```python
nips = pd.read_csv('../../../Data/NIPS_1987-2015.csv')
nips
```

```python
nips_linkage = linkage(nips.iloc[:,1:], 'ward')
nips_linkage.shape
```

```python
plt.figure(figsize=(35,20))
dendrogram(nips_linkage, orientation='top')
plt.show()
```

While it is hard to interpret the dendogram, we can still make some observations:
    - Among the the 2 main clusters, one of them captures a vast majority of the data points. (green)
    - Within the green cluster, most sub groups of data points get clustered immediately with their distance score being under 1000 (mostly in the range of 0-300).
    - Within the red cluster, it takes more distance to merge the sub groups within.
    - The highest distance between sub groups in red and green cluster is around 2500.
